```python
import numpy as np
L = 15.7122

N = 216 #int(input("Entter the number of atoms :"))
N1 = 120 #int(input("Entter the number of 1st type of atom(Si) :"))  
cut_off_S = 2.6 #= float(input("Enter the cutoff length for Si-Si bonding: "))
cut_off_O = 1.7 #= float(input("Enter the cutoff length for Si-Si bonding: "))

data_points_raw = np.loadtxt("SPC_SiOx.vasp", skiprows = 8, max_rows = N)# loads the file and keeps needed coordinates
data_points_S = data_points_raw[:N1] #keeps only first atom type coordinates
data_points_O = data_points_raw[N1:N]
data_points_t_S = np.transpose(data_points_S) # transpose to 3XN1 arrays. did this so we can work on arrays of same coordinates type.
data_points_t_O = np.transpose(data_points_O)

x_S = data_points_t_S[0] ## stores X-coordinates of all Si atoms
y_S = data_points_t_S[1]
z_S = data_points_t_S[2]

x_O = data_points_t_O[0]# stores X-coordinates of all O atoms
y_O = data_points_t_O[1]
z_O = data_points_t_O[2]

#print(np.shape(data_points_t))
#print(data_points_t[0])
#number = 17 ## looking for particular Si - atom


def BC(dx):  ## satisfying the boundary conditions.
    dx_bound = np.where(dx > L/2, 1, 0) # creates an array that has 1 for those dx > L/2; boundary condition imposed
    return abs((L*dx_bound - dx)) ## returns an array of distance that satisfies BC
    
    
def get_results(dx, dy, dz, cut_off): #computes the distance between the atoms coordinates and also check for the
                                       # condition of cutoff for each bondining of atoms.
    bond_length_2 = (BC(dx))**2 + (BC(dy))**2 + (BC(dz))**2
    cond_satis = np.where(bond_length_2 < (cut_off**2)) ## stores the atomic position of those atoms that satisfies
                                                        # condition to atoms to lie in a sphere of that atom.
    return cond_satis[0][cond_satis[0] != i], bond_length_2 # returns atomic osition and corres[ponding bond length]
    
for i in range(N1):## checking only for Si atoms as these are one that bonds with either Si or O atom.
    dx_S = L*abs(x_S-x_S[i]) ; dy_S = L*abs(y_S-y_S[i]); dz_S = L*abs(z_S-z_S[i])
    dx_O = L*abs(x_O-x_S[i]) ; dy_O = L*abs(y_O-y_S[i]); dz_O = L*abs(z_O-z_S[i])

    final_S, bond_length_S = get_results(dx_S, dy_S, dz_S, cut_off_S)## Si atoms position satifying the condition
    final_O, bond_length_O = get_results(dx_O, dy_O, dz_O, cut_off_O)## O atoms 
    
    print()
    #$ final_S=1 and i+1 for atoms position starting from 1. otherwise 0 for first atom
    print(f'{i+1} Si-atom = {final_S+1}Si-atom, Bond-lengths=  {np.sqrt(bond_length_S[final_S])}.')
    print()
    print(f'{i+1} Si-atom = {final_O+1+N1}O-atom, Bond-lengths = {np.sqrt(bond_length_O[final_O])}.')
    print()
    
    # to get for particular silicon atom
    #if i == number-1:
    #   print(final_S+1, final_O+N1+1)
        
```
